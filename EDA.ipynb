```python
fig , ax = plt.subplots(2, 2, figsize=(15,10))

SubscriptionStatus = df.groupby("Subscription Status")["Purchase Amount (USD)"].sum().reset_index().sort_values(by="Purchase Amount (USD)", ascending=False)
PaymentMethod = df.groupby("Payment Method")["Purchase Amount (USD)"].sum().reset_index().sort_values(by="Purchase Amount (USD)", ascending=False)
ShippingType = df.groupby("Shipping Type")["Purchase Amount (USD)"].sum().reset_index().sort_values(by="Purchase Amount (USD)", ascending=False)

# Plot 1
sns.barplot(x="Subscription Status", y="Purchase Amount (USD)", data=SubscriptionStatus, color = "red" , ax=ax[0,0])
ax[0,0].set_title("Total Revenue by Subscription Status")

# Plot 2
sns.barplot(x="Payment Method", y="Purchase Amount (USD)", data=PaymentMethod, color = "yellow" , ax=ax[0,1])
ax[0,1].set_title("Total Revenue  by Payment Method")

# Plot 3
sns.barplot(x="Shipping Type", y="Purchase Amount (USD)", data=ShippingType, color = "blue" , ax=ax[1,0])
ax[1,0].set_title("Total Revenue by Shipping Type")

ax[1,1].axis('off')

plt.tight_layout()
plt.show()
```
